Pandas Import

```python
import pandas as pd
```

Daten von URL laden

```python
df = pd.read_csv('https://raw.githubusercontent.com/edlich/eternalrepo/master/DS-WAHLFACH/dsm-beuth-edl-demodata-dirty.csv')
```

```python
df
```

Die Spalten "id" (damit man ansonsten doppelte Zeilen komplett entfernen kann) und "full_name" (der redundant ist) entfernen.

```python
df.drop(['id', 'full_name'], inplace=True, axis=1)
```

Doppelte Einträge entfernen, dabei wollen wir alle Spalten miteinbeziehend (geben also "subset" nicht mit an) und diese direkt im df entfernen (inplace=true) und den ersten Eintrag der zu den Doppelten gehört behalten (geben also "keep" nicht mit an, da Default "first" ist)

```python
df.drop_duplicates(inplace=True)
```

Alter "age" nur positive (".abs()") Ints zulassen

```python
df['age'] = pd.to_numeric(df['age'], errors='coerce').abs()
```

Fehlende Angaben ersetzen wo sie sinnvoll ersetzbar sind wie bei gender wegen anderer Identifikation der Person

```python
df['gender'] = df.gender.fillna('other')
```

Zuletzt alle Teilnehmer mit fehlenden Angaben entfernen

```python
df.dropna(inplace=True)
```

Als weiteres Beispiel nur Volljährige behalten, z.B. falls Daten für eine Studie

```python
df = df[df.age > 18]
```
